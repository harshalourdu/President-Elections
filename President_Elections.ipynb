```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
election_data = pd.read_csv("election2024.csv")
```

```python
print("Dataset Information:")
election_data.info()
```

```python
print("\nDescriptive Statistics:")
print(election_data.describe())
```

```python
print("\nMissing Values:")
print(election_data.isnull().sum())
```

```python
print("\nGender Distribution:")
print(election_data['sex'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(election_data['age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
print("\nFederal District Distribution:")
print(election_data['federal_district'].value_counts())
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=election_data, x='will_vote', order=election_data['will_vote'].value_counts().index)
plt.title('Voting Intention')
plt.xlabel('Will Vote')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=election_data, x='candidate', order=election_data['candidate'].value_counts().index)
plt.title('Candidate Preference')
plt.xlabel('Candidate')
plt.ylabel('Count')
plt.show()
```

```python
print("\nEducation Level Distribution:")
print(election_data['education'].value_counts())
```

```python
print("\nIncome Level Distribution:")
print(election_data['income'].value_counts())
```

```python
print("\nEmployment Status Distribution:")
print(election_data['employment'].value_counts())
```

```python
print("\nTelevision Usage Frequency:")
print(election_data['television_usage'].value_counts())
```

```python
print("\nInternet Usage Frequency:")
print(election_data['internet_usage'].value_counts())
```

```python
plots = [
    ('age_distribution.png', plt.figure(figsize=(10, 6)), sns.histplot(election_data['age'].dropna(), kde=True)),
    ('voting_intention.png', plt.figure(figsize=(10, 6)), sns.countplot(data=election_data, x='will_vote', order=election_data['will_vote'].value_counts().index)),
    ('candidate_preference.png', plt.figure(figsize=(10, 6)), sns.countplot(data=election_data, x='candidate', order=election_data['candidate'].value_counts().index))
]

for filename, fig, plot in plots:
    plot.set_title(filename.split('.')[0].replace('_', ' ').title())
    fig.savefig(filename)
    plt.close(fig)
```
